```python
# -------------------------------------------------------------------------------
#
#                              ğŸ“šğŸ“šğŸ“šğŸ“š
#
# -------------------------------------------------------------------------------
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
# -------------------------------------------------------------------------------
#
#                              ğŸ¤ğŸ¤ğŸ¤ğŸ¤
#
# -------------------------------------------------------------------------------
#
#                              ğŸ–¥ï¸ğŸ“
#
# -------------------------------------------------------------------------------
try:
  driver = webdriver.Chrome()
except:
  driver = webdriver.Firefox()
# driver.get method() will navigate to a page given by the URL address
driver.get('https://www.linkedin.com/login')
#locate email form by_class_name
username = driver.find_element(By.ID, 'username')
# send keys(0) to simulate keystrokes
username.send_keys(usero)
# sleep for 0.5 seconds
print("ğŸ˜´ğŸ˜´: ğŸ§")
sleep(0.5)
# locate password form by_class_name
password = driver.find_element(By.ID,'password')
# send keys() to simulate key strokes
password.send_keys(passwordo)
print("ğŸ˜´ğŸ˜´: ğŸ”‘")
sleep(0.5)
# locate submit button by xpath
sign_in_button = driver.find_element(By.XPATH,'//* [@type="submit"]')
# . click() to mimic button click
sign_in_button.click()
print("ğŸ˜´ğŸ˜´: ğŸƒ")
sleep(10)
# -------------------------------------------------------------------------------
#
#                              ğŸ•µï¸ğŸ•µï¸ğŸ•µï¸ğŸ•µï¸
#
# -------------------------------------------------------------------------------
# locate the search bar element
search_field = driver.find_element(By.XPATH, '//*[@id="global-nav-typeahead"]/input')
# input the search query to the search bar
search_field.send_keys(lookfor)
# search
search_field.send_keys(Keys.RETURN)
print("ğŸ˜´ğŸ˜´: ğŸ•µï¸")
sleep(5)
# locate Peoples button by xpath
print("âš ï¸ linkedin sometimes changes the order of these buttons!")
peoplebttn = driver.find_element(By.XPATH,'//*[@id="search-reusables__filters-bar"]/ul/li['+str(buttonnum)+']/button')
# click it
peoplebttn.click()
print("ğŸ˜´ğŸ˜´: ğŸƒ")
sleep(7)
# -------------------------------------------------------------------------------
#
#                              ğŸ•µï¸ğŸ•µï¸ğŸ•µï¸ğŸ•µï¸
#
# -------------------------------------------------------------------------------
profilesurl = []
pythonisweird=1

for ii in range(0,numofpages):
  print(f" I just read page # {ii+1:d}")
  profiles = driver.find_elements(By.XPATH, '//div[@class="display-flex"]')
  for profile in profiles:
    try:
      profile = profile.find_element(By.CLASS_NAME,classname)
      profile = profile.get_attribute('href')
      if profile not in profilesurl:
        profilesurl.append(profile)
    except:
      pythonisweird=0
      print(" ğŸ˜â— didnt find anything")
  # locate next peoples button by xpath
  driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
  print("ğŸ˜´ğŸ˜´: ğŸ‘‡")
  sleep(3)
  nextbttn = driver.find_element(By.XPATH,'//button[@aria-label="Next"]')
  # click it
  nextbttn.click()
  sleep(2)
# -------------------------------------------------------------------------------
#
#                              â›”â˜ï¸â›”â˜ï¸â›”
#
# -------------------------------------------------------------------------------
driver.quit()
```
